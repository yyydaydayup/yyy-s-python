```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 20)
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
```

```python
values = np.arange(20)

plt.subplot(221)
plt.plot(values)
plt.plot(values**2)  # axisplot

plt.subplot(222)
plt.plot(np.sin(values),'r')

plt.subplot(223)
plt.plot(np.cos(values), 'g')

plt.subplot(224)
plt.plot(np.tan(values), 'b')
```

```python
values = np.linspace(0,20)
plt.style.use('seaborn')
fig, axes = plt.subplots(2, 2, sharex=True)
# figaxesaxis
ax1 = axes[0, 0]
ax1.plot(values)

ax2 = axes[0, 1]
ax2.plot(np.sin(values), 'r')

ax3 = axes[1, 0]
ax3.plot(np.cos(values), 'g')

ax4 = axes[1, 1]
ax4.plot(np.tan(values), 'b')
```
